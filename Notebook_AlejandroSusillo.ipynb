<h2>Data modeling</h2>

This section will apply several classification algorithms to the dataset. We are looking for those algorithms with the best accuracy metric.

```python
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn import neighbors
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.decomposition import PCA
import xgboost as xgb
# to create confusion matrix
from sklearn.metrics import accuracy_score, confusion_matrix, precision_recall_fscore_support
import matplotlib.pyplot as plt
import seaborn as sns
# AVOID WARNING MESSAGES
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import time
```

```python
from sklearn.preprocessing import LabelEncoder

df_aux = df.copy(deep=True)
labelencoder=LabelEncoder()
for col in df_aux.columns:
    df_aux[col] = labelencoder.fit_transform(df_aux[col])
df_aux.head()
```

```python
cols = ['cap-shape','cap-surface','cap-color','odor','gill-color','stalk-root','stalk-surface-above-ring','stalk-surface-below-ring','stalk-color-above-ring','stalk-color-below-ring','veil-color','ring-number','ring-type','spore-print-color','population','habitat']

df_num = pd.get_dummies(df[cols])
df_num = df_aux.join(df_num)
df_num = df_num.drop(cols,axis=1)
df_num.head()
```

```python
x = df_num.iloc[:,1:len(df_num)]
y = df_num.iloc[:, 0]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1)
```

<h3>Decision Trees</h3>

We'll use 70% of the dataset, and for the second one a 30%.

Let's try a Hold-Out evaluation first:

```python
clf=DecisionTreeClassifier()
clf=clf.fit(x_train, y_train)
y_pred=clf.predict(x_test)
acc=accuracy_score(y_pred, y_test)
print('Tree Accuracy by hold-out evaluation: ',acc)
```

As we can see the accuracy for the algorithm is 100% which is unreal, so let's try this time N-fold Cross validation.

```python
clf=DecisionTreeClassifier()
acc_nfold_dt = cross_val_score(clf, x, y, cv=5, scoring='accuracy').mean()

start_kfold_dt = time.clock()
y_pred = cross_val_predict(clf, x, y, cv=5)
end_kfold_dt = time.clock() - start_kfold_dt
kfold_time_dt = end_kfold_dt

print("Tree Accuracy by N-fold Cross Validation:", acc_nfold_dt)
```

We have an accurary of 93%, which is closer to reality. What has happened before is because the dataset may be too smail to perform a Hold-Out evaluation.

```python
get_conf_matrix(y, y_pred)
```

The confusion matrix compares those values returned by the algorithm with the real values that the algorithm should have returned. In case the algorithm returns a 'edible' value for the class variable for a specific mushroom and that mushroom was actually edible, we'll have one instance more in the top right corner of the confusion matrix. Otherwise, if the algorithm classified it as poisonous and it was actually poisonous, it will be in the bottom right corner. 

You'll notice that there are two more corners remaining. The top one is for what is called 'false positives'. That means that the algorithm classified that instance as true but it was actually false. The opposite happens with the bottom left corner ('false negative').

<h3>Association Rules</h3>

```python
#from apyori import apriori
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df_rules = df_knn[relevant_features.index]

df_aux = pd.get_dummies(df[['class','bruises','stalk-shape']])
df_rules = df_rules.join(df_aux)
df_rules = df_rules.drop(['class','bruises','stalk-shape'],axis=1)
df_rules.head()
```

We have got the top-10 rules sorted by lift that have 3 or less antecedents and which consequent is class, edible or poisonous.

```python
frequent_itemsets = apriori(df_rules, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.sort_values(by='lift', inplace=True, ascending=False)
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules = rules[rules['antecedent_len'] <= 4]
rules_p = rules[(rules['consequents'] == {'class_p'})]
rules_e = rules[(rules['consequents'] == {'class_e'})]
```

```python
def get_conf_matrix(y_real, y_pred):
    # Creates a confusion matrix
    cm = confusion_matrix(y_real, y_pred) 
    # Transform to df for easier plotting
    cm_df = pd.DataFrame(cm,
                         index = ['0','1'], 
                         columns = ['0','1'])
    print(cm_df)

    plt.figure(figsize=(5.5,4))
    sns.heatmap(cm_df, annot=True)
    plt.title('Confusion Matrix \nAccuracy:{0:.2f}'.format(accuracy_score(y_real, y_pred)))
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
def get_pie_chart(colX, colY):
    from matplotlib.gridspec import GridSpec
    import matplotlib.gridspec as gridspec

    fig, axes = plt.subplots(figsize=(18,8))
    gs = gridspec.GridSpec(1, 3)

    for i, x in zip(range(0, len(list(df.iloc[:, colX:colY]))), list(df.iloc[:, colX:colY])):
        plt.subplot(gs[0, i])
        sizes = []
        explode = []
        uniques = df[x].unique()
        labels = list(uniques)
        colors = ['orange', 'aquamarine', 'antiquewhite','chocolate', 'darkgray', 'yellow', 'purple', 'lightcoral', 'green', 'darkorchid', 'forestgreen', 'goldenrod', 'greenyellow']

        for e in uniques:
            sizes.append(len(df[df[x]==e]))
            explode.append(0)

        plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
        plt.axis('equal')
    plt.show()
```
